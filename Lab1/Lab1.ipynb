```python
import pandas as pd
```

```python
uname = ['UserID', 'gender', 'age', 'occuoation', 'Zip-code']
```

```python
users = pd.read_table("users.dat", sep = "::", engine ='python', names = uname)
```

```python
users
```

```python
mov_col = ['MovieID', 'Titel', 'Genres']
```

```python
movies = pd.read_table("movies.dat", sep = "::", engine = 'python', names = mov_col, encoding = "ISO-8859-1")
```

```python
movies
```

```python
rat_col = ['UserID','MovieID','Rating','Timestamp']
```

```python
ratings = pd.read_table("ratings.dat", sep = "::", engine = 'python', names = rat_col)
```

```python
ratings
```

```python
dataFrame = ratings.merge(users, how = 'inner', on = uname[0]).merge(movies, how = 'inner', on = mov_col[0])
```

```python
under_18_Group = dataFrame[dataFrame['age'] == 1];
between_18_24_Group = dataFrame[dataFrame['age'] == 18];
between_25_34_Group = dataFrame[dataFrame['age'] == 25];
between_35_44_Group = dataFrame[dataFrame['age'] == 35];
between_45_49_Group = dataFrame[dataFrame['age'] == 45];
between_50_55_Group = dataFrame[dataFrame['age'] == 50];
older_56_Group = dataFrame[dataFrame['age'] == 56];
```
